```python
#--- 2025-08-15 12-12  by Dr. Thawatchai Chomsiri
import numpy as np
import os

# 
activation_files = [
    "loss_000_ReLU6.npz",
    "loss_000_GELU.npz",
    "loss_000_PARALU.npz",
    "loss_000_LSGELU.npz",
    "loss_000_XSINELU.npz",
    "loss_000_XSINELU2Pi.npz",
    "loss_000_ELU.npz",
    "loss_000_Swish.npz",
    "loss_000_ReLU.npz",
    "loss_000_SECHLU.npz",
        
]

# Dictionary  loss  activation function
combined_loss = {}

#  loss 
for file in activation_files:
    if os.path.exists(file):
        with np.load(file) as data:
            # Extract the activation name from the filename
            activation_name = file.split('_')[2].replace('.npz', '')
            # Store the loss
            combined_loss[activation_name] = data['loss']

#  combined 
np.savez("loss_000.npz", **combined_loss)

print(f"Combined loss saved to loss_000.npz: {combined_loss}")
```
